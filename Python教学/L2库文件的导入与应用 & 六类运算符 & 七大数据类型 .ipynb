# 库文件的导入与应用
    - 什么是库文件
        - 库文件为第三方封装包 用于实现各种不同功能
    - 怎么安装库文件
        - pip3 install name
    - 怎么导入库文件
        - import name
        - import name as short_name
        - from name import func
        - from name import func as short_func
        - from name import *
    - 怎么调用库文件
        - name.func()
        - short_name.func()
        - func()
        - short_func()
        - func()

```python
import math
print( math.pow(2,3) )
import math as ma
print( ma.pow(2,3) )
from math import pow
print( pow(2,3) )
from math import pow as p 
print( p( 2,3 ) )
from math import *
print( pow(2,3) )

# test 
from pprint import pprint as p
```

 - list
   - list 的 拼接
   - list 的 索引 和 切片
   - list 的 增删查改

```python
lst = [1,2,3]
print(lst)
# 增
lst_add = lst + [4,5] # 合并两个list，按顺序
lst_mul = lst * 2 # 将一个list的内容copy指定次数 并组成一个新的list
lst.append(4) # 在list末尾添加指定数据
lst.extend((666,777)) # 在末尾添加很多个值，数据格式为iterable即可
lst.insert(0,0) # insert(index, item) 在指定index添加item
print(lst_add,lst_mul,lst)
# 删
lst.pop(0) # 删除指定index的数据，index不存在则报错, 返回删除掉的内容
lst.remove(1) # 删除在list中正序找到的第一个匹配的指定数据
print(lst)
# 查
print(len(lst),max(lst),min(lst),lst.count(3),lst.index(3))
# 改
lst.reverse() # 将所有元素顺序反过来
lst.copy()  # 创建一个一模一样但是内存地址不一样的lst
lst.sort(reverse= True) # 升序排序,改变原来的list
sorted(lst)  # 升序排序，不改变原来的list，创建一个新list
```
