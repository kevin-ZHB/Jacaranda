**操作性函数**
- return None
- 对原有的输入进行改变

```python
ori = []
def change_ori(ori):
    ori.append('i am changed')

print(change_ori(ori))
print(ori)

def name(some):
    return sorted(some)
```

**输出性函数**
- return required output
- 对原有输入不改变

```python
ori = []
def keep_ori(ori):
    ori = ori.copy()
    ori.append('i am changed')
    return ori

print(keep_ori(ori))
print(ori)
print('\n')
```

# 参数
1.  必须参数(must)
    <br>Def function(name):
    <br>    调用函数的时候必须输入的参数, name
2. 默认参数(dafult)
    <br>Def function(name=Jacaranda):
    <br>调用函数的时候可以选择不输入的参数，不输入时为默认值

```python
def canshu1(must,default = '默认'):
    print('这是必须参数',must)
    print('这是默认参数',default,'\n')

canshu1('a',default='b')
canshu1('a')
```

3. 不定长参数
    <br>调用函数的时候可以输入无限个数据的参数
    
    

Def function(*name):
- *name的数据会以tuple的形式导入
  - 如1,2导入后为（1,2）    

```python
def canshu2(must,*long):
    print('这是必须参数',must)
    print('这是不定长参数的tuple导入形式',long)

canshu2('a','b','c')
```

```python
def canshu2(must, must_1, *long):
    print('这是必须参数',must,must_1)
    print('这是不定长参数的tuple导入形式',long)

canshu2('a','b','c')
```

Def function(**name):
- *name的数据会以dictionary的形式导入
  - 如a=1,b=2导入后为 {a:1, b:2}

```python
def canshu3(must,**long):
    print('这是必须参数',must)
    print('这是这是不定长参数的字典导入形式',long)

canshu3('a',b='b1',c='c2')
```

**全局变量 (global variable)**
- 有效范围为整个程序空间

**局部变量 (local variable)**
- 函数内部定义的变量
-  出了函数就不能使用了

```python
all = '我是全局变量'
def get_all():
    #global all
    all = '我是局部变量'
    print(all)
get_all()
print(all)
```

**Lambda**
- Lambda 是一个匿名函数，即只包括函数的 process, 但是不包括函数的名字
- 是一种简单的、在同一行中定义函数的方法
- lambda \<参数>: \<return value>

```python
def my_key(item):
    return item[::-1]

my_list = ['cz', 'bz', 'az']
print(sorted(my_list, key=my_key))
print(sorted(my_list, key=lambda item: item[::-1]))
```

# 类 Class
1.  用来描述具有相同属性和方法的对象的集合，它定义了集合中每一个对象所共有的
属性和方法
2. 当我们需要对一类事情进行统一化重复性操作的时候，为了减少工作量，我们会选
择创建一个 class 来将他们归类，从而避免对这一类对象的共有属性进行多次编辑，
从而更注重于他们的不同

**类的创建**
<br>

Class（家长）
> Attribute属性（子孙）-遗传家长特征，如长相，身高，体型
>> Methods方法 （生活技能） -继承生活技能，如说话，行走，跑步

```python
class people:
    height = 180
    look = ' like 吴彦祖'
    age  = 18 
    def __init__(self,name,age = 18,look = 'good',height = 180):
        self.name = name
        self.age = age 
        self.look = look
        self.height = height 
        self.uglies = self.uglies(people)
    def say_hi(self):
        print('hello i am '+self.name)
        self.intro()
    def intro(self):
        print('i am {}'.format(self.age))
    class uglies:
        def __init__(self,out):
            self.out = out
            print(out.height)
wuyanzu = people('吴彦祖',age = 20)
wuyanzu.say_hi()
print(wuyanzu)

```

**Methods 方法**
<br>某个物件可用来执行特定任务的函式

```python
class GC: 
    level = 1
    exp = 0
    
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def attack(self):
        print(f"{self.name} attacked!")
        
    def watch(self):
        print(f"{self.name} stares at you O.o")

player1 = GC("john", 18)
player2= GC("doe", 22)

player1.attack()
player2.watch()
```

**私有方法**
<br>只能内部调用

```python
class GC: 
    level = 1
    exp = 0
    
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def __speak(self): #类的私有方法
        print('AHAHAHAHAHAHAHAHAHAHAHA')
        
    def yell(self): #调用私有方法
        self.__speak()
        
player1 = GC("john", 18)
player2= GC("doe", 22)


player1.yell()
```

**类对象的继承**
<br>一个新的类可以继承一个父类的所有方法和属性
<br>

```python
class GC: 
    level = 1
    exp = 0
    
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def attack(self):
        print(f"{self.name} attacked!")
        
    def watch(self):
        print(f"{self.name} stares at you O.o")
        
class hop(GC):
    def jump(self):
        print(f"{self.name} jumpped!!!")

player1 = hop("john doe", 99)
player1.attack()
player1.jump()
```
